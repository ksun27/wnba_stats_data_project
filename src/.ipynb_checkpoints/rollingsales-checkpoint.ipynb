## **6. describe the kind of buildings that were sold**

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

d = {
  "One Family Homes": 0,
  "Office Buildings": 0,
  "Condominiums": 0,
  "Other": 0
}
x = df['BUILDING CLASS AT TIME OF SALE'].to_list()
for ele in x:
    if ele.startswith('A'):
        d["One Family Homes"] += 1
    elif ele.startswith('O'):
        d["Office Buildings"] += 1
    elif ele.startswith('R'):
        d["Condominiums"] += 1
    else:
        d["Other"] += 1

x = d.keys()
y = d.values()
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.bar(x, y, align='center', color='#aaddff')
plt.ylabel('Number Sold')
plt.xlabel('Building Type')
plt.title('Units Sold Per Type')
```

I turned the data from the column into a list that I could then use a dictionary to put data into. I used a for loops with conditional statements to add values to my dictionary. I then plotted the dictionary with corresponding keys and values and turned it into a bar graph.

## **8. bin the prices of properties sold**

```python
groups = [0, 500000, 2000000, 10000000, 1000000000, 2500000000]
a = saleprice.groupby(pd.cut(saleprice, groups)).count()
print(a)
a.plot.bar()
```

I binned it by groups I thought fit the data. The dataset's prices are very heavily skewed around the 500,000-2,000,000 price range and that there were several outliers especially on the high price side. These outliers can skew other statistics from earlier calculations. A good way to fix this is to remove the outliers that are outside 2 standard deviations from the mean.

## **9. create a visualization the shows the relationship (if any) between the price sold and the number of square footage of all the areas of a property within a building or a structure**

```python
df['GROSS SQUARE FEET'].dropna(how=any)
c = df['GROSS SQUARE FEET'].str.replace(',', '').astype(float)
sqft = c.values
prices = saleprice.values
graph = {"Square Feet": sqft, "Prices": prices}
graphdf = pd.DataFrame(data=graph)
graphdf.plot.scatter(x="Square Feet", y="Prices")
```

## **10. which month did the least amount of sales occur**

```python
time = df['SALE DATE'].to_list()
sales = saleprice.to_list()
salesPerMonth = {"January": 0, "February": 0, "March": 0, "April": 0, "May": 0, "June": 0, "July": 0, "August": 0, "September": 0, "October": 0, "November": 0, "December": 0}
for i in range(len(time)):
    if time[i].startswith('1'):
        salesPerMonth["January"] += sales[i]
    if time[i].startswith('2'):
        salesPerMonth["February"] += sales[i]
    if time[i].startswith('3'):
        salesPerMonth["March"] += sales[i]
    if time[i].startswith('4'):
        salesPerMonth["April"] += sales[i]
    if time[i].startswith('5'):
        salesPerMonth["May"] += sales[i]
    if time[i].startswith('6'):
        salesPerMonth["June"] += sales[i]
    if time[i].startswith('7'):
        salesPerMonth["July"] += sales[i]
    if time[i].startswith('8'):
        salesPerMonth["August"] += sales[i]
    if time[i].startswith('9'):
        salesPerMonth["September"] += sales[i]
    if time[i].startswith("10"):
        salesPerMonth["October"] += sales[i]
    if time[i].startswith("11"):
        salesPerMonth["November"] += sales[i]
    if time[i].startswith("12"):
        salesPerMonth["December"] += sales[i]
leastSalesMonth = min(salesPerMonth, key=salesPerMonth.get)
print(salesPerMonth)
print(leastSalesMonth)
```

To find the month with the least sales (price-wise), I used a to_list() function to clean up both the times and sales data. I then made a dictionary that kept track of sales with the corresponding month. Using a for loop, I then aggregated the sales to their months and then used a min() function to find the month with the least sales. I can conclude that sales reach a lull around the spring time with March being the month with the least sales.
